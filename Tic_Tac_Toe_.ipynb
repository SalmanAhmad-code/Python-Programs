```python
# Function to print the Tic Tac Toe board
def print_tic_tac_toe(values):
    print("\n")
    print("\t     |     |")
    print(f"\t  {values[0]}  |  {values[1]}  |  {values[2]}")
    print("\t_____|_____|_____")
    print("\t     |     |")
    print(f"\t  {values[3]}  |  {values[4]}  |  {values[5]}")
    print("\t_____|_____|_____")
    print("\t     |     |")
    print(f"\t  {values[6]}  |  {values[7]}  |  {values[8]}")
    print("\t     |     |")
    print("\n")

# Function to print the scoreboard
def print_scoreboard(score_board):
    print("\t--------------------------------")
    print("\t          SCOREBOARD            ")
    print("\t--------------------------------")
    for player, score in score_board.items():
        print(f"\t  {player} \t {score}")
    print("\t--------------------------------\n")

# Function to check if a player has won
def check_win(player_pos, cur_player):
    winning_combos = [
        [1, 2, 3], [4, 5, 6], [7, 8, 9],    # Rows
        [1, 4, 7], [2, 5, 8], [3, 6, 9],    # Columns
        [1, 5, 9], [3, 5, 7]                # Diagonals
    ]
    for combo in winning_combos:
        if all(pos in player_pos[cur_player] for pos in combo):
            return True
    return False

# Function to check if the game is a draw
def check_draw(player_pos):
    return len(player_pos['X']) + len(player_pos['O']) == 9

# Function to play a single game
def single_game(cur_player):
    values = [' '] * 9
    player_pos = {'X': [], 'O': []}

    while True:
        print_tic_tac_toe(values)
        try:
            move = int(input(f"Player {cur_player}, choose your box (1-9): "))
        except ValueError:
            print("Invalid input! Please enter a number between 1 and 9.")
            continue

        if move < 1 or move > 9:
            print("Invalid move! Try again.")
            continue

        if values[move - 1] != ' ':
            print("Box already filled! Choose another.")
            continue

        # Update board and positions
        values[move - 1] = cur_player
        player_pos[cur_player].append(move)

        # Check for win
        if check_win(player_pos, cur_player):
            print_tic_tac_toe(values)
            print(f"Player {cur_player} wins!\n")
            return cur_player

        # Check for draw
        if check_draw(player_pos):
            print_tic_tac_toe(values)
            print("It's a draw!\n")
            return 'D'

        # Switch player
        cur_player = 'O' if cur_player == 'X' else 'X'

# Main game execution
if __name__ == "__main__":
    print("Player 1")
    player1 = input("Enter name: ")
    print("\nPlayer 2")
    player2 = input("Enter name: ")
    print()

    cur_player = player1
    player_choice = {'X': "", 'O': ""}
    options = ['X', 'O']
    score_board = {player1: 0, player2: 0}

    print_scoreboard(score_board)

    while True:
        print(f"{cur_player}, choose your symbol:")
        print("1. X")
        print("2. O")
        print("3. Quit")

        try:
            choice = int(input("Enter choice: "))
        except ValueError:
            print("Invalid input! Please enter 1, 2, or 3.\n")
            continue

        if choice == 1:
            player_choice['X'] = cur_player
            player_choice['O'] = player2 if cur_player == player1 else player1
        elif choice == 2:
            player_choice['O'] = cur_player
            player_choice['X'] = player2 if cur_player == player1 else player1
        elif choice == 3:
            print("Final Scores:")
            print_scoreboard(score_board)
            break
        else:
            print("Invalid choice! Please try again.\n")
            continue

        winner = single_game(options[choice - 1])

        if winner != 'D':
            winning_player = player_choice[winner]
            score_board[winning_player] += 1

        print_scoreboard(score_board)
        cur_player = player2 if cur_player == player1 else player1
```
